### Задание 2. Проведите исследование полученных функций. 

1. $5x + 1.5$
2. $2e^{0.2x}-1$
3. $3\sqrt x - 15$
4. $2\log(x) + 5$
5. $-5x^2 - 10x + 100$




```python
from sympy import *
from sympy.solvers.inequalities import solve_univariate_inequality
init_printing(use_unicode=False,wrap_line=False,no_global=True)
```

```python
x = Symbol('x')
one = 5*x+1.5
plot(one)
print(one.subs(x,100))
print(one.subs(x,-100))
print('Функция не четная')
print('Нули функции',solve(one))
print('Больше нуля ',solve_univariate_inequality(one>0,one))
print('Меньше нуля',solve_univariate_inequality(one<0,one))
print('Возрастает на всем участке')
```

```python
three = 3*sqrt(x)-15
plot(three)
print('Функция не четная')
print('Нули функции',solve(three))
```

```python
four = 2*log(x) +5 
plot(four)
print('Функция не четная')
print('Нули функции',solve(four))
```

```python
five = -5*x**2 - 10*x + 100
plot(five)
print(five.subs(x,100))
print(five.subs(x,-100))
print('Функция не четная, но симметричная')
print('Нули функции',solve(five))
```
