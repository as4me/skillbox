```python
from sympy import *
from sympy.plotting import *
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x , y')
f1 = x*(y-5)**2
f2 = (y-exp(1)/20)**2*sin(pi*x)
f3 = x*sin(x)*sin(y-2)

f1x = diff(f1,x)
f1y = diff(f1,y)

f2x = diff(f2,x)
f2y = diff(f2,y)

f3x = diff(f3,x)
f3y = diff(f3,y)
```

```python
print(f1x, " | ", f1y)
print(f2x, " | ", f2y)
print(f3x, " | ", f3y)
```

Задание 2

f(x1,x2) = a2x2+a1x1+a0 <br>
1 авто   x1 = 1945 x2 = 560 y1 = 4.3 <br>
2 авто   x1 = 1495 x2 = 340 y1 = 4.9 <br>
3 авто   x1 = 1570 x2 = 343 y1 = 5.2 <br>

```python
a1, a2, a0 = symbols('a1, a2, a0')
MSE = 1/3*(((a2 * 560 + a1 * 1945 + a0)-4.3)**2 + \
           ((a2 * 340 + a1 * 1495 + a0)-4.9)**2 + \
           ((a2 * 343 + a1 * 1570 + a0)-5.2)**2)
MSE
```

```python
MSEa2 = diff(MSE,a2)
MSEa2
```

```python
MSEa1 = diff(MSE,a1)
MSEa1
```

```python
MSEa0 = diff(MSE,a0)
MSEa0
```

```python
nonlinsolve([MSEa2,MSEa1,MSEa0],[a2,a1,a0])
```

```python
MSE.subs({a2: -.0118811881188066, a1:0.00447524752474972,a0:2.24910891089351})
```
