Обучите модель DBSCAN с параметром *min_samples=8*. Визуализируйте результат.

```python
import pickle
import numpy as np

with open('7.10._clustering.pkl','rb') as f:
    data_clustering = pickle.load(f)
    
data_clustering

X = np.array(data_clustering['X'])
Y = np.array(data_clustering['Y'])
```

```python
from sklearn.cluster import DBSCAN

db =DBSCAN(eps=0.6, min_samples=8).fit(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0], X[:,1], s=40, c=db.labels_, marker='o', alpha=0.8, label='data')
```

```python
db =DBSCAN(eps=0.5, min_samples=8).fit(X)
plt.scatter(X[:,0], X[:,1], s=40, c=db.labels_, marker='o', alpha=0.8, label='data')
```

```python
db =DBSCAN(eps=0.4, min_samples=8).fit(X)
plt.scatter(X[:,0], X[:,1], s=40, c=db.labels_, marker='o', alpha=0.8, label='data')
```
